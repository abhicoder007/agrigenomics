### Attribute Information: (classes: edible=e, poisonous=p)

```python
import numpy as np
import pandas as pd
df=pd.read_csv("mushrooms.csv")
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
X=df.iloc[:,1:]
Y=df.iloc[:,0]
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
d=defaultdict(LabelEncoder)
Xfit=X.apply(lambda x: d[x.name].fit_transform(x))
```

```python
Xfit.shape
```

```python
Xfit
```

```python
le_y=LabelEncoder()
yfit=le_y.fit_transform(Y)
```

```python
yfit.shape
```

```python
Xfit.drop_duplicates(inplace=True)
```

```python
Xfit.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(Xfit,yfit,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
Logreg=LogisticRegression()
Logreg.fit(X_train,Y_train)
y_pred=Logreg.predict(X_test)
```

```python
y_pred=Logreg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("accuarcy",metrics.accuracy_score(Y_test,y_pred))
```

```python
print("precison",metrics.precision_score(Y_test,y_pred))
```
